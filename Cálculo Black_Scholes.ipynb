```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros
S = np.linspace(50, 150, 100)  # Variao do preo do ativo
K = 100  # Preo de exerccio
T = 1  # Tempo at o vencimento em anos
r = 0.05  # Taxa de juros livre de risco
 = 0.2  # Volatilidade

# Frmulas de d1 e d2
d1 = (np.log(S / K) + (r + **2 / 2) * T) / ( * np.sqrt(T))
d2 = d1 -  * np.sqrt(T)

# Preo da opo de venda
P = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

# Plotar o grfico
plt.figure(figsize=(10, 6))
plt.plot(S, P, label='Preo da Opo de Venda')
plt.xlabel('Preo do Ativo Subjacente (S)')
plt.ylabel('Preo da Opo de Venda (P)')
plt.title('Preo da Opo de Venda usando o Modelo Black-Scholes')
plt.legend()
plt.grid(True)
plt.show()
```
